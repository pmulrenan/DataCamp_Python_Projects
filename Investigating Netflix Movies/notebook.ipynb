## The data

| Column | Description |
|--------|-------------|

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Most frequent movie duration in the 1990s
netflix_movies = netflix_df[netflix_df['type'] == 'Movie']
netflix_90s = netflix_movies[(netflix_movies['release_year'] >= 1990) & (netflix_movies['release_year'] <2000)]
netflix_90s.hist('duration', bins = 20, color = 'darkred', edgecolor = 'black')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (Mins)')
plt.ylabel('Number of Movies')
plt.grid(False)
plt.show()
```

```python
# 90s Actions Movies Under 90 Minutes
netflix_90s_action = netflix_90s[netflix_90s['genre'] == 'Action']

short_movie_count = 0

for label, row in netflix_90s_action.iterrows() :
    if row['duration'] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count

print(short_movie_count)
```
